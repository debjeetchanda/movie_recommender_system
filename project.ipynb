```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title') #merge both dataframes
```

```python
movies.info()
```

```python
movies = movies[['movie_id', 'keywords', 'genres', 'title', 'cast', 'crew', 'overview','release_date']]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
#converting release date into year as year only matters more

movies.dropna(inplace=True)
movies['release_date']= movies['release_date'].apply(lambda x: int(x[0:4]))
#lets categorise the release date into the decade they were released as people might prefer a certain decade films more
movies['release_date']= movies['release_date'].apply(lambda x: (x//10)*10)
```

this is clearly a string of list of dictionaries
first we have to convest this to a list
['Action','Adventure','Fantasy']

```python
#to convert to string we have to use the function literal_eval from the module ast
import ast
ast.literal_eval(movies.iloc[0].genres)
```

```python
a = movies.iloc[0].genres
b = []
for i in ast.literal_eval(movies.iloc[0].genres):
    b.append(i['name'])
b
```

```python
#we have to convert this to a function which takes input as string 
def convert(a):
    b = []
    for i in ast.literal_eval(a):
        b.append(i['name'])
    return b
    
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

WE are checking keywords column as well now

```python
movies.iloc[0].keywords
#has the same problem so we use same function
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
#we have to create a similar function for this so for that we can create convert3 function as we have to find top 3 cast in each movie
def convert3(a):
    b = []
    counter = 0        
    for i in ast.literal_eval(a):
        if counter !=3:                 #this counter will help in picking only the top 3 or the first 3 iterations in the list of distionaries
            b.append(i['name'])
        else:
            break
    return b
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies.head(5)
```

```python
ast.literal_eval(movies['crew'][0])
```

```python
#we have to only extract the name where the job is director from the list of dictionaries
def convertdir(a):
    b = []
    for i in ast.literal_eval(a):
        if i['job'] == "Director":
            b.append(i['name'])
            break
    return b
convertdir(movies['crew'][0])
```

```python
movies['crew'] = movies['crew'].apply(convertdir) #apply to the whole column
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#since this is a string we convert this to list
movies['overview'] = movies['overview'].apply(lambda x :x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
#we are removing spaces within the string elements itself in genres,cast, crew and keywords features as often this leads to confusion in the model as 
# for example we have a Sam worthington and a Sam mendes so there is a confusion between  both if we treat sam and worthington or mendes separately
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+ movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
newdf = movies[['movie_id','title','tags','release_date']]
```

```python
newdf['tags'] = newdf['tags'].apply(lambda x:" ".join(x))
```

```python
newdf.head()
```

```python
#convert every string to lower
newdf['tags'] = newdf['tags'].apply(lambda x: x.lower())
```

```python
newdf.head()
```

we are stemming the data

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def helper(x):
    return " ".join([ps.stem(word) for word in x.split()])
newdf['tags'] = newdf['tags'].apply(helper)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english', max_features=5000)
vectors = cv.fit_transform(newdf['tags']).toarray()
```

```python
vectors[0]
```

```python
newdf
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
np.argsort(similarity[0])[::-1][0:5]
```

```python
newdf.iloc[[1,3,5,6],1]
```

```python
def recommend(movie,n=5):
    if movie not in np.array(newdf['title']):
        return 'No such movie exists in our database'
    if n>50:
        n = 50
    ind = newdf[newdf['title']==movie].index[0]
    v = list(np.argsort(similarity[ind])[::-1][0:51])
    c = list(np.sort(similarity[ind])[::-1][0:51])
    di = dict(zip(v,c))
    M = pd.DataFrame(list(di.items()), columns=['index', 'similarityvalue'])
    M['release_date'] = newdf.iloc[v,3]
    def datesimilarity(x):
        if (x-M.iloc[0,2])==0:                 #we are handling 0 values here like this
            return 100
        else:
            return 1/(x-M.iloc[0,2])
    M['release_date'] = M['release_date'].apply(datesimilarity)
    M['netsimilarity'] = M['similarityvalue']**2 + 0.01*(M['release_date']**2)
    final = M.sort_values(by='netsimilarity', ascending=False)[1:n+1]
    fin = list(final['index'])
    return newdf.iloc[fin]
#this will find 5 most similar movies
```

```python
recommend('Titanic',10)
```
